1. Cities and towns (points and polygons) – .shp / .shx / .dbf / .prj / .cpg — векторные геоданные (точки и полигоны населённых пунктов) в формате ESRI Shapefile.

2. Roads and railroads – .shp / .shx / .dbf / .prj / .cpg — векторные линейные объекты (дороги и железные дороги) в формате ESRI Shapefile.

3. Airports and ports – .shp / .shx / .dbf / .prj / .cpg — векторные точечные объекты (аэропорты и порты) в формате ESRI Shapefile.

4. Power plants – .csv (+ .pdf / .txt как документация) — табличные данные по электростанциям (атрибуты станций, координаты и характеристики) в обычном CSV.

5. Gridded Population 2015 (250 m) – .tif / .tif.ovr / .tif.aux.xml / .tif.xml — растровый слой плотности населения в формате GeoTIFF (ячейки 250 м).

6. Gridded Gross Domestic Product and Human Development Index over 1990–2015 – .nc — многомерные растровые поля в формате NetCDF (GDP, HDI по времени и пространству).

7. Gridded Land cover 2015 – .tif — растровый слой типов землепользования (классы покрытия поверхности) в формате GeoTIFF.

8. Tree Cover Loss by Dominant Driver – .tif / .tif.ovr / .tif.aux.xml / .tif.xml — растровый слой потери древесного покрова по типу драйвера в формате GeoTIFF.

9. Carbon accumulation potential from natural forest regrowth in forest and savanna biomes – .tif — растровый слой потенциала аккумулирования углерода (поглощение при естественном возобновлении леса) в формате GeoTIFF.

10. Solar energy potential – .tif / .tif.pdf / .tif.xml — растровые слои солнечного потенциала (GHI, GTI, PVOUT) в формате GeoTIFF.

11. Air temperature – .tif / .tif.pdf / .tif.xml — растровый слой температуры воздуха (долгосрочное среднее) в формате GeoTIFF.

12. Global cattle distribution in 2010 (5 minutes of arc) – .tif — растровый слой распределения поголовья крупного рогатого скота (разрешение 5 arc‑minutes) в формате GeoTIFF.



1. **1_CITIES_landscan**  
   - `ne_10m_populated_places/*` — shapefile с точками населённых пунктов: координаты городов, атрибуты (название, страна, население, другие поля).
   - `ne_10m_urban_areas_landscan/*` — shapefile с полигонами городских / урбанизированных территорий (контуры агломераций, городских зон).
   → То есть отдельно: точки городов и полигоны урбанизированных областей.

2. **2_Roads_railroads**  
   - `ne_10m_railroads/*` — shapefile с линейными объектами железных дорог (геометрия линий + атрибуты вроде типа дороги).
   - `ne_10m_roads/*` — shapefile с автомобильными дорогами (магистрали, основные дороги и т.п.).
   → Две векторные сети: железные дороги и автодороги.

3. **3_Airports_ports**  
   - `ne_10m_airports/*` — shapefile с точками аэропортов (координаты + атрибуты: название, код и т.п.).
   - `ne_10m_ports/*` — shapefile с точками морских/речных портов (координаты + атрибуты).
   → Точечная инфраструктура: по отдельным слоям для аэропортов и портов.

4. **4_globalpowerplantdatabasev120**  
   - `global_power_plant_database.csv` — табличный набор по всем электростанциям: местоположение, тип генерации, мощность, статус и т.д.
   - `README.txt`, `RELEASE_NOTES.txt`, `A_Global_Database_of_Power_Plants.pdf` — документация и описание методологии.
   → Один крупный CSV + документация — это не карты, а атрибутивные данные с координатами.

5. **5_GHS_POP_E2015_GLOBE_R2019A_54009_250_V1_0**  
   - `..._2015_...250_V1_0.tif` (+ `.ovr`, `.xml`) — растровая карта населения за 2015 год, 250 м; в каждой ячейке значение численности/плотности населения.
   - `GHSL_Data_Package_2019_light.pdf` — документация по продукту GHSL (что именно за метрики, единицы измерения).

6. **6_GDP/doi_10.5061_dryad.dk1j0__v2**  
   - `GDP_PPP_1990_2015_5arcmin_v2.nc` — временной ряд GDP (PPP) на сетке 5 arc‑min.
   - `GDP_PPP_30arcsec_v3.nc` — более детальный GDP (PPP) на 30 arc‑sec.
   - `GDP_per_capita_PPP_1990_2015_v2.nc` — GDP per capita PPP во времени.
   - `HDI_1990_2015_v2.nc` — HDI (индекс развития) во времени. geospatial-environmental-and-socioeconomic-data)
   - `admin_areas_GDP_HDI.nc` — агрегированные GDP/HDI по административным единицам.
   - `pedigree_*.nc` — «pedigree» файлы с информацией о качестве/происхождении оценок GDP и HDI.
   - папка `kummu_etal_scidata_code` (`*.m`, `Readme.rtf`) — код MATLAB и описание того, как авторы собрали и обработали данные.

7. **7_Land_cover**  
   - `ESACCI-LC-L4-LCCS-Map-300m-P1Y-2015-v2.0.7.tif` — глобальная карта типов землепокрова (land cover) на 2015 год, 300 м; значения ячеек — классы LCCS (лес, вода, сельхозземли и т.п.).

8. **8_GFW_deforestation**  
   - `Goode_FinalClassification_19_05pcnt_prj.tif` (+ `.ovr`, `.xml`) — растровый слой потери лесного покрова с классификацией по доминирующему драйверу (аграрная экспансия, вырубка, пожары и т.д.), подготовленный Global Forest Watch.

9. **9_GFW_potential_carbon_sequestration**  
   - `young_forest_sequestration_rate_Griscom_extent.tif` — растр потенциальных скоростей аккумулирования углерода при естественном возобновлении леса (молодой лес) в лесных и саванных биомах.

10. **10_solar_potential**  
   - `World_GHI_.../GHI.tif` (+ `.pdf`, `.xml`) — глобальная карта GHI (Global Horizontal Irradiance, суммарная солнечная радиация на горизонтальную поверхность), типично среднесуточная/суммарная.
   - `World_GTI_.../GTI.tif` (+ `.pdf`, `.xml`) — глобальная карта GTI (Global Tilted Irradiance, на наклонную поверхность).
   - `World_PVOUT_.../PVOUT.tif` (+ `.pdf`, `.xml`) — глобальная карта PVOUT (ожидаемая выработка от PV‑панелей при заданных предположениях).
   → Три разных показателя солнечного потенциала, каждый со своим описанием в PDF/XML.

11. **11_temperature**  
   - `World_TEMP_.../TEMP.tif` (+ `.pdf`, `.xml`) — глобальная карта температуры воздуха (обычно долгосрочные средние или климатические показатели), геопривязанный растр.
12. **12_Cattle**  
   - `1_Ct_2010_Metadata.html` — описание и метаданные по глобальному распределению крупного рогатого скота.
   - `6_Ct_2010_Aw.tif` — растровая карта распределения поголовья крупного рогатого скота (значения в ячейках = плотность / количество голов на заданную площадь).



растровые данные - хранятся в виде сетки пикселей (ячеек), а не как точки. Прямоугольная матрица, где у каждой ячейки свое значение (геокоордината + значение высоты/температуры/плотности и тд). Формат GeoTIFF.

векторные данные - описывают геометрию объекта (точки городов, линии объектов и тд) и их атрибуты. Хранятся в формате shp.

```python
from rasterio.windows import Window
import matplotlib.pyplot as plt
```

```python
print(temperature.width, temperature.height)  # чтобы понять размер

# допустим, хотим 1000x1000 пикселей из центра
w, h = 1000, 1000
row_off = (temperature.height - h) // 2
col_off = (temperature.width - w) // 2

window = Window(col_off=col_off, row_off=row_off, width=w, height=h)

data = temperature.read(1, window=window)

plt.figure(figsize=(6, 6))
plt.imshow(data, cmap="coolwarm")
plt.colorbar(label="Temperature")
plt.title("Temperature (window)")
plt.axis("off")
plt.show()
```
